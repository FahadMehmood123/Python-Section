```python
key=input("Enter key to put in matrix: ")
get_key=""
for i in key:
    get_key=get_key+i.lower()

get=int(input("\n1.Encryption\n2.Decryption: \nYour choice: "))  
lst=[]

fix=[[0 for i in range(5)] for j in range(5)]
for i in get_key:
    if i not in lst:
        if i!='j':
            lst.append(i)
        else:
            lst.append('i')
            
count=0
for i in range(97,123):
    if chr(i) not in lst:
        if i==105 and chr(106) not in lst:
            lst.append("i")
            count=1
        elif count==0 and i==105 or i==106:
            pass    
        else:
            lst.append(chr(i))


temp2=0
for i in range(0,5):
    for j in range(0,5):
        fix[i][j]=lst[temp2]
        temp2=temp2+1

def location(let):
    if let=='j':
        let='i'
    lst2=[]
    for i ,j in enumerate(fix):
        for k,l in enumerate(j):
            if let==l:
                lst2.append(i)
                lst2.append(k)
                return lst2
            
if get==1:
    temp=input("Enter message: ")
    msg=""
    for i in temp:
        if i==" ":
            i=""
            msg=msg+i.lower()
        else:
            msg=msg+i.lower()
            
    for temp in range(0,len(msg)+1,2):
        if temp<len(msg)-1:
            if msg[temp]==msg[temp+1]:
                msg=msg[:temp+1]+'x'+msg[temp+1:]        
    if len(msg)%2!=0:
        msg=msg[:]+'x'
        
    print("\n Encrypted text: ",end=' ')
    i=0
    while i<len(msg):
        let=[]   
        let1=[]
        let=location(msg[i])
        let1=location(msg[i+1])
        if let[1]==let1[1]:
            print(fix[(let[0]+1)%5][let[1]],end='')
            print(fix[(let1[0]+1)%5][let1[1]],end=' ')
            
        elif let[0]==let1[0]:
            print(fix[let[0]][(let[1]+1)%5],end='')
            print(fix[let1[0]][(let1[1]+1)%5],end=' ')  
            
        else:
            print(fix[let[0]][let1[1]],end='')
            print(fix[let1[0]][let[1]],end=' ')    
        i=i+2        
                 
elif get==2:
    temp=input("Enter Encrypted Text: ")
    msg=""
    for i in temp:
        if i==" ":
            i=""
            msg=msg+i.lower()
        else:
            msg=msg+i.lower()
    print("\n Original Message: ",end=' ')
    
    i=0
    while i<len(msg):
        let=[]
        let1=[]
        let=location(msg[i])
        let1=location(msg[i+1])
        if let[1]==let1[1]:
            print(fix[(let[0]-1)%5][let[1]],end='')
            print(fix[(let1[0]-1)%5][let1[1]],end=' ')
            
        elif let[0]==let1[0]:
            print(fix[let[0]][(let[1]-1)%5],end='')
            print(fix[let1[0]][(let1[1]-1)%5],end=' ')  
            
        else:
            print(fix[let[0]][let1[1]],end='')
            print(fix[let1[0]][let[1]],end=' ')    
        i=i+2
```
