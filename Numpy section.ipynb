# Dot Product vs Elemnnt wise

```python
np.random.seed(0)
mat1=np.random.randint(10,size=(5,3))
mat2=np.random.randint(10,size=(5,3))
mat2
```

```python
mat1*mat2 #Element wise so it will follow the rule of broadcasting
```

```python
mat3=np.dot(mat1,mat2.T) #it will follow the rule of matrix multiplication
mat3
```
