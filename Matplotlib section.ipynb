# Histogram

```python
fig,axe=plt.subplots()
axe.set(title="Histogram-plot")
axe.hist(x**2);
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(13,11))
ax1.set(title="line-graph",xlabel="x-axis",ylabel='y-axis')
ax1.plot(x,x**2);
ax2.set(title="scatter-plot",xlabel="x-axis",ylabel="yaxis")
ax2.scatter(x,np.sin(x));
ax3.set(title="histogram",xlabel="x-axis",ylabel='y-axis')
prac=np.linspace(0,10,100)
ax3.hist(prac**2);
dic={"Fruit-salad":50,
    "Yougurt":110,
    "Milk":100}
ax4.set(title="Bar-Graph",xlabel="x-axis",ylabel='y-axis')
ax4.bar(dic.keys(),dic.values());
```

```python
fig,ax =plt.subplots(nrows=2,ncols=2,figsize=(13,11))
ax[0,0].plot(x,x**2);
ax[0,0].set(title='Line-Graph')
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[0,1].set(title="Scatter-Graph")
ax[1,0].bar(dic.keys(),dic.values());
ax[1,0].set(title="Bar-Graph")
ax[1,1].hist(prac**2);
ax[1,1].set(title="Histogram");
```

# OO Method(Object Oriented)
*This method is use for advanced plot and has more features than pyplot method

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
       ax=ax);

ax.set_xlim([45,100]);
```

```python
plt.style.use("seaborn-whitegrid")
fig,ax=plt.subplots(figsize=(10,6))
var=ax.scatter(x=over_50['age'],
              y=over_50['chol'],
              c=over_50['target'],
              cmap="winter")
ax.set(title="Age against Heart Disease",
      xlabel="Age",
      ylabel="Cholestrol",
      )
ax.legend(*var.legend_elements(),title="Target");
ax.axhline(over_50['chol'].mean(), color='r', linestyle='--');
plt.xlim([50,80])
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,9));

var1=ax[0].scatter(x=over_50['age'],
                   y=over_50['chol'],
                  c=over_50['target'],
                   cmap="winter")

var2=ax[1].scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target'],
                   cmap='winter')

ax[0].set(title='Heart Disease and Cholestrol Level',
         ylabel='Choletrol')
ax[1].set(title="Heart Disease and Max Heart Disease Level",
         ylabel='Max Heart Rate')

ax[0].legend(*var1.legend_elements(),title="Target",fontsize=10);
ax[1].legend(*var2.legend_elements(),title="Target");

ax[1].axhline(over_50['thalach'].mean(), color='r', linestyle='--');
ax[0].axhline(over_50['chol'].mean(), color='r', linestyle='--');

ax[1].set_ylim([60,200]);  
ax[1].set_xlim([50,80]);
ax[0].set_xlim([50,80]);

fig.suptitle(t='Heart Disease Analysis',fontsize=16,fontweight="bold");
```
